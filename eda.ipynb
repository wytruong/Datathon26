```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from scipy import stats
```

```python
def plot_dist(ax, data, col, color, label, log=False):
    """Helper: histogram + KDE + summary stats on a single axis."""
    x = data[col].dropna()
    if log:
        x = np.log1p(x)
        xlabel = f"log1p({col})"
    else:
        xlabel = col

    ax.hist(x, bins=40, color=color, alpha=0.55, density=True, edgecolor="white", linewidth=0.4)
    kde_x = np.linspace(x.min(), x.max(), 300)
    kde = stats.gaussian_kde(x)
    ax.plot(kde_x, kde(kde_x), color=color, linewidth=2)
    ax.axvline(x.mean(),   color="black",  linestyle="--", linewidth=1.2, label=f"Mean  {x.mean():.2f}")
    ax.axvline(x.median(), color="dimgray", linestyle=":",  linewidth=1.2, label=f"Median {x.median():.2f}")

    ax.set_title(label, fontsize=12, fontweight="bold")
    ax.set_xlabel(xlabel, fontsize=9)
    ax.set_ylabel("Density", fontsize=9)
    ax.legend(fontsize=8)
    ax.spines[["top", "right"]].set_visible(False)

palette = ["#4C72B0", "#DD8452", "#55A868", "#C44E52", "#8172B2"]
```

```python
fig = plt.figure(figsize=(18, 14))
fig.suptitle("Section 2 — Univariate Distributions (facility-year level)", fontsize=15)
gs = gridspec.GridSpec(3, 3, figure=fig, hspace=0.55, wspace=0.35)

ax0 = fig.add_subplot(gs[0, 0])
ax1 = fig.add_subplot(gs[0, 1])
ax2 = fig.add_subplot(gs[0, 2])
plot_dist(ax0, df, "ed_visit",      palette[0], "ED Visits (log-scale)", log=True)
plot_dist(ax1, df, "ed_admit",      palette[1], "ED Admissions (log-scale)", log=True)
plot_dist(ax2, df, "burden_score",  palette[2], "Burden Score")
```

```python
fig = plt.figure(figsize=(18, 14))

ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])
ax5 = fig.add_subplot(gs[1, 2])
plot_dist(ax3, df, "true_burden_score", palette[3], "True Burden Score (log-scale)", log=True)
plot_dist(ax4, df, "ed_admit",    palette[4], "Admission Rate (admit/visit)")
plot_dist(ax5, df[df["has_treatment_data"] == 1], "treatment_stations", palette[0],
          "Treatment Stations (log-scale)", log=True)
```

```python
ax6 = fig.add_subplot(gs[2, 0])
ax7 = fig.add_subplot(gs[2, 1:])   # wider

# Service level bar chart
sl_counts = df["er_service_level_desc"].value_counts()
ax6.barh(sl_counts.index, sl_counts.values, color=palette[:len(sl_counts)], edgecolor="white")
ax6.set_title("ER Service Level — Facility Count", fontsize=12, fontweight="bold")
ax6.set_xlabel("# Facilities")
ax6.spines[["top", "right"]].set_visible(False)
for i, v in enumerate(sl_counts.values):
    ax6.text(v + 1, i, str(v), va="center", fontsize=8)

# Top 20 counties by facility count
county_counts = df.groupby("county_name")["oshpd_id"].nunique().sort_values(ascending=False).head(20)
ax7.bar(county_counts.index, county_counts.values, color=palette[1], edgecolor="white", alpha=0.85)
ax7.set_title("Top 20 Counties — Unique ED Facilities", fontsize=12, fontweight="bold")
ax7.set_xlabel("County")
ax7.set_ylabel("# Facilities")
ax7.tick_params(axis="x", rotation=45, labelsize=8)
ax7.spines[["top", "right"]].set_visible(False)

plt.tight_layout()
```
