```python
import numpy as np        
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns    
```

```python
df = pd.read_csv("data/ca_ed_final.csv")
print(f"Shape: {df.shape}")
print(f"Dtypes: {df.dtypes}")
print(f"Null counts: {df.isnull().sum()}")
```

```python
# Drop rows missing core metrics (ed_visit, burden_score, true_burden_score)
df = df.dropna(subset=['ed_visit', 'burden_score', 'true_burden_score'])

# Fill missing ed_admit with 0 (visit occurred but no admits recorded)
df['ed_admit'] = df['ed_admit'].fillna(0)

print(f"Shape after cleaning: {df.shape}")
print(f"Remaining nulls:\n{df.isnull().sum()}")
```

```python
df = df.dropna(subset=['burden_score_normalized'])
```
